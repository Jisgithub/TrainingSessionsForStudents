```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np

# Load the MNIST dataset
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()

# Preprocess the data (rescale to 0-1)
x_test = x_test / 255.0

# Build and train a simple model for demonstration
# In your actual code, you would have a trained model
model = tf.keras.models.Sequential([
    tf.keras.layers.Flatten(input_shape=(28, 28)),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(10, activation='softmax')
])
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
# A simple way to "train" for this example, you would use x_train and y_train
model.fit(x_test, y_test, epochs=1)

# Make predictions
predictions = model.predict(x_test)
predicted_labels = tf.argmax(predictions, axis=1)

# Now, visualize the first 10 images with their true and predicted labels
num_images_to_show = 10
fig, axes = plt.subplots(1, num_images_to_show, figsize=(20, 3))

for i in range(num_images_to_show):
    ax = axes[i]
    ax.imshow(x_test[i], cmap='gray')

    true_label = y_test[i]
    predicted_label = predicted_labels[i].numpy()

    # Set the title with both the true and predicted labels
    # Use green for a correct prediction, red for an incorrect one
    title_color = 'green' if true_label == predicted_label else 'red'
    ax.set_title(f"True: {true_label}\nPred: {predicted_label}", color=title_color)

    ax.axis('off') # Hide the axes

plt.tight_layout()
plt.show()
```
